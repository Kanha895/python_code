```python
def prompt_code(text):
    prompt=f"""You are an Java Code generator which generates java code for the problem statement generated by the user....
    the problem statement is as follows : {text}
    Keep the code easy to understand for the user and also keep the code little complex that can make code readable....
    """
    response=llm.invoke(prompt)
    return response
    
```

```python
print("Enter a problem statement whose java code you want to generate (type 'exit' to quit):")
while True:
    user_input=input("Enter your Problem Statement:")

    if user_input.lower()=="exit":
        print("Exiting")
        break
    else:
        result = prompt_code(user_input)
        print(f"Result:\n{result.content}\n")
```
